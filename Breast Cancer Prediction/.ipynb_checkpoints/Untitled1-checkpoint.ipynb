```python
import numpy as np
import pandas as pd
import seaborn as sns 
from sklearn.preprocessing import StandardScaler, LabelEncoder 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score,accuracy_score
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_excel("C:\\wine_quality.xlsx")
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.dropna(subset=["pH","fixed acidity","citric acid","volatile acidity","residual sugar","chlorides","pH","sulphates"], inplace=True)
```

```python
data.isnull().sum()
```

```python
data
```

```python
data.duplicated().sum()
```

```python
data = data.drop_duplicates()
```

```python
data.duplicated().sum()
```

```python
data['quality'] = data['quality'].astype('category',copy=False)
```

```python
data.info()
```

```python
data.describe()
```

```python
label_encoder = LabelEncoder()

data['type'] = label_encoder.fit_transform(data['type'])  
data['type']
```

```python
data.value_counts("type")
```

```python
X = data.drop(columns="quality")
y = data['quality']
X
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
y
```

```python
data['quality'].value_counts()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=50)
```

```python
lr = LogisticRegression(multi_class="multinomial", solver="lbfgs", max_iter=1000)
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
print(f"Accuracy: {accuracy: .2f}")
```
